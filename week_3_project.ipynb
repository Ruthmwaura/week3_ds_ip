**DATA LOADING**

```python
#loading the geo data on the environment 
Geo = pd.read_csv('/content/cells_geo.csv', delimiter=';', index_col=0)
Geo.head()
#using the provided description we rename the column
Geo.columns=['City','Status','Locality','Geographical_Zone','Zone_Name','Longitude','Latitude','Region','Area','Cell_ID','Site_ID']
Geo.head()
```

```python
#loading CDR data to the environment
#DAY 1
#CDR 20120507
CDR= pd.read_csv('http://bit.ly/Telcom_dataset1', delimiter=',')
CDR.head()
#Renaming the columns with understanding from description
CDR.columns=['Product','Billing_Price','Date_Time','Cell_On_Site','Party_A_Number','Party_B_Number','Country_A','Country_B','Cell_ID','Site_ID']
CDR.head()
```

```python
#CDR 20120508
#DAY 2
CDR1= pd.read_csv('http://bit.ly/Telcom_dataset2', delimiter=',')
CDR1.head()
#Renaming columns
CDR1.columns=['Product','Billing_Price','Date_Time','Cell_On_Site','Party_A_Number','Party_B_Number','Country_A','Country_B','Cell_ID','Site_ID']
CDR1.head()
```

```python
#CDR 20120509
#DAY 3
CDR2= pd.read_csv('http://bit.ly/Telcom_dataset3', delimiter=',')
CDR2.head()
#renaming the columns
CDR2.columns=['Product','Billing_Price','Date_Time','Cell_On_Site','Party_A_Number','Party_B_Number','Country_A','Country_B','Cell_ID','Site_ID']
CDR2.head()
```

**DATA** **CLEANING**


```python
#JOINING THE THREE CDR DATASET FOR THE THREE DAYS
CDR_joined = pd.concat([CDR, CDR1, CDR2], ignore_index = True)
CDR_joined
```

```python
#Spilliting date and time to have two columns
CDR_joined [['Date', 'Time']] = CDR_joined.Date_Time.str.split(expand=True)
CDR_joined.head()
```

```python
#DROPPING IRRELEVANT COLUMNS
#WE WILL NOT NEED THE COUNTRIES  DATE TIME AND CELL ON SITE
CDR_joined.drop(columns= ['Country_A', 'Date_Time' ,'Country_B', 'Cell_On_Site'], inplace=True)
CDR_joined.head()
```

```python
#DROPPING THE NULL VALUES IN THE CDR DATASET
CDR_joined= CDR_joined.dropna()
CDR_joined.head()
```

```python
#DROPPING THE DUPLICATES
CDR_joined= CDR_joined.drop_duplicates()
CDR_joined.head()
```

```python
#Reviewing the data after the "droppings"
CDR_joined
#2214 rows have been dropped
```

```python
#cleaning geo data
#dropping rows with null values
Geo= Geo.dropna()
Geo.head()
#dropping duplicates
Geo= Geo.drop_duplicates()
Geo
```

```python
#joining CDR and geo data
#using SITE id which is in both datasets
CDR_Geo = Geo.merge(CDR_joined, how='left', on='Site_ID')
CDR_Geo.head()
```

```python
#DROPPING IRRELEVANT COLUMNS
CDR_Geo.drop(columns= ['Longitude', 'Latitude' ,'Party_A_Number', 'Party_B_Number'], inplace=True)
CDR_Geo.head()
```
